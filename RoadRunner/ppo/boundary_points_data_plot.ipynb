```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# ICML Settings

```python
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times", "Times New Roman"], 
    "font.size": 10,
    "axes.labelsize": 11,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
})
```

```python
df = pd.read_csv('boundary_points.csv')
feature_cols = [f'Feature_{i}' for i in range(1, 17)]
df_top10 = df.head(10)

actions = df_top10['Action_Label'].values
custom_x_labels = [f"S{i+1}\n(Act {int(a)})" for i, a in enumerate(actions)]

df_heatmap = df_top10[feature_cols].T
df_heatmap.columns = custom_x_labels
df_heatmap.index = [f"$F_{{{i+1}}}$" for i in range(16)]

fig, ax = plt.subplots(figsize=(10, 7))

sns.heatmap(df_heatmap, 
            annot=True, 
            fmt=".2f", 
            cmap="YlOrRd", 
            linewidths=.75, 
            cbar_kws={'label': 'Feature Value', 'shrink': 0.8},
            annot_kws={"size": 8.5},
            ax=ax)

ax.set_xlabel("State Index & Corresponding Action ($s, a$)", fontsize=11, labelpad=10)
ax.set_ylabel("State Features ($R^{16}$)", fontsize=11)
plt.xticks(rotation=0) 

plt.savefig("RoadRunner_ppo_boundary_points_heatmap.pdf", dpi=600, bbox_inches='tight')
plt.show()
```
